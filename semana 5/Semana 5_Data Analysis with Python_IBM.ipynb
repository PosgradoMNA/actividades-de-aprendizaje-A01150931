```python
mean_X12 = df['X12'].mean()
df['X12'].replace(np.nan, mean_X12,  inplace = True)
df.loc[15:22 , 'X10':'X19']
```

```python
#To check the type of each column
df.info()
```

```python
'Converting the first column from int to float'
df['X1'].astype(float)
```

```python
#Normalize a column
df['X15'] = (df['X15']-df['X15'].mean()) / df['X15'].std()
df
```

```python
#Creating bins
bins = np.linspace(min(df['X1']), max(df['X1']),4)
groups = ['Low', 'Medium', 'High']

df['Grupos de precios'] = pd.cut(df['X1'],bins, labels=groups, include_lowest=True)
df
```
